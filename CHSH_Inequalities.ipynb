1. Let $A, B, C$ be random binary variables. **Prove that the following inequality holds**
$$
\mbox{Pr}[A = 1 \wedge B = 0] + \mbox{Pr}[B = 1 \wedge C = 0] \ge \mbox{Pr}[A = 1 \wedge C = 0].
$$

For convenience, denote $A = 1$ as $\mathcal{A}$, $A = 0$ as $\overline{\mathcal{A}}$, and same to $B$ and $C$, the target inequality becomes $$\mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{B}}\right] + \mbox{Pr}\left[\mathcal{B} \wedge \overline{\mathcal{C}}\right] \ge \mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{C}}\right]$$
Since probabilities are non negative, we have: $$
\mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{B}} \wedge \mathcal{C}\right] + \mbox{Pr}\left[\overline{\mathcal{A}} \wedge \mathcal{B} \wedge \overline{\mathcal{C}}\right] \ge 0.
$$
By adding $\mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{C}}\right]$ to both sides:
\begin{align*}
&\mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{B}} \wedge \mathcal{C}\right] + \mbox{Pr}\left[\overline{\mathcal{A}} \wedge \mathcal{B} \wedge \overline{\mathcal{C}}\right] + \mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{C}}\right] \\
=\ &\mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{B}} \wedge \mathcal{C}\right] + \mbox{Pr}\left[\overline{\mathcal{A}} \wedge \mathcal{B} \wedge \overline{\mathcal{C}}\right] + \mbox{Pr}\left[\mathcal{A} \wedge \mathcal{B} \wedge \overline{\mathcal{C}}\right] + \mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{B}} \wedge \overline{\mathcal{C}}\right]\\
=\ &\left(\mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{B}} \wedge \mathcal{C}\right]+\mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{B}} \wedge \overline{\mathcal{C}}\right]\right)+\left(\mbox{Pr}\left[\overline{\mathcal{A}} \wedge \mathcal{B} \wedge \overline{\mathcal{C}}\right] + \mbox{Pr}\left[\mathcal{A} \wedge \mathcal{B} \wedge \overline{\mathcal{C}}\right]\right) \\
=\ &\mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{B}}\right] + \mbox{Pr}\left[\mathcal{B} \wedge \overline{\mathcal{C}}\right]\\
\ge\ &\mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{C}}\right]
\end{align*}

2. Using the previous inequality, **prove the following inequality holds**

$$
\mbox{Pr}[A \neq B] + \mbox{Pr}[B \neq C] \ge \mbox{Pr}[A \neq C].
$$

By using same denotion as question 1, we have $$\mbox{Pr}\left[A \neq B\right] = \mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{B}}\right] + \mbox{Pr}\left[\overline{\mathcal{A}} \wedge \mathcal{B}\right],\ \mbox{Pr}[B \neq C]= \mbox{Pr}\left[\mathcal{B} \wedge \overline{\mathcal{C}}\right] + \mbox{Pr}\left[\overline{\mathcal{B}} \wedge \mathcal{C}\right],\ \mbox{Pr}[A \neq C]=\mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{C}}\right] + \mbox{Pr}\left[\overline{\mathcal{A}} \wedge \mathcal{C}\right]$$
From the conclusion of question 1, 
\begin{equation}
\mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{B}}\right] + \mbox{Pr}\left[\mathcal{B} \wedge \overline{\mathcal{C}}\right] \ge \mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{C}}\right] \tag{1}
\end{equation}
Similarly, by adding $\mbox{Pr}\left[\overline{\mathcal{A}} \wedge \mathcal{C}\right]$ to both sides of $\mbox{Pr}\left[\overline{\mathcal{A}} \wedge \mathcal{B} \wedge \overline{\mathcal{C}}\right] + \mbox{Pr}\left[\mathcal{A} \wedge \overline{\mathcal{B}} \wedge \mathcal{C}\right] \ge 0$,
\begin{equation}
\mbox{Pr}\left[\overline{\mathcal{A}} \wedge \mathcal{B}\right] + \mbox{Pr}\left[\mathcal{\overline{B}} \wedge \mathcal{C}\right] \ge \mbox{Pr}\left[\overline{\mathcal{A}} \wedge \mathcal{C}\right] \tag{2}
\end{equation}
By adding (1) and (2), we have $\mbox{Pr}\left[A \neq B\right]+ \mbox{Pr}[B \neq C] \ge \mbox{Pr}[A \neq C]$.

$$
\mbox{Prob}[A_2 \neq B_2] - \mbox{Prob}[A_1 \neq B_1] - \mbox{Prob}[A_2 \neq B_1] - \mbox{Prob}[A_1 \neq B_2] \le 0.
$$

If all the quantum states obeys local realism, it means that

(1) $A_1$, $A_2$, $B_1$, $B_2$ are independent

(2) One's measurements do not affect the other one's measurements

From the conclusion of question 2, we have 
\begin{align*}
&\mbox{Prob}\left[A_1 \neq B_1 \right] + \mbox{Prob}\left[A_2 \neq B_1 \right] + \mbox{Pr}\left[A_1 \neq B_2 \right] \\
\ge\ &\mbox{Prob}\left[A_1 \neq A_2 \right] + \mbox{Pr}\left[A_1 \neq B_2 \right] \\
\ge\ &\mbox{Prob}\left[A_2 \neq B_2 \right]
\end{align*}
which is exactly $\mbox{Prob}[A_2 \neq B_2] - \mbox{Prob}[A_1 \neq B_1] - \mbox{Prob}[A_2 \neq B_1] - \mbox{Prob}[A_1 \neq B_2] \le 0$.

4. Let $X_i$ and $Y_j$ be random binary variables such that $X_i = (-1)^{A_i}$ and $Y_j = (-1)^{B_j}$, respectively, for $i,j \in \{1,2\}$. Let $E[X_iY_j]$ be the expected values of the multiplication of random variables $X_i$ and $Y_j$. Then, by using the above CHSH inequality, **prove the following** (which is a different form of CHSH inequality on the expectation of random variables). 
$$
E[X_1 Y_1] + E[X_1 Y_2] + E[X_2 Y_1] - E[X_2 Y_2] \le 2 
$$

Denote $\Lambda$ as the measure space of $A_1, A_2, B_1, B_2$, 
\begin{align*}
&E\left[X_1 Y_1\right] + E\left[X_1 Y_2\right] + E\left[X_2 Y_1 \right] - E\left[X_2 Y_2\right] \\
=\ &\sum\nolimits_{\Lambda}P_{\Lambda}\cdot X_1 Y_1 + \sum\nolimits_{\Lambda}P_{\Lambda}\cdot X_1 Y_2 + \sum\nolimits_{\Lambda}P_{\Lambda}\cdot X_2 Y_1 -\sum\nolimits_{\Lambda}P_{\Lambda}\cdot X_2 Y_2\\
=\ &\sum\nolimits_{\Lambda}P_{\Lambda}\cdot\left[X_1(Y_1+Y_2)+X_2(Y_1-Y_2)\right] \tag{3}
\end{align*}
Let $T(X_1, X_2, Y_1, Y_2) = X_1(Y_1+Y_2)+X_2(Y_1-Y_2)$

1. When $B_1=B_2$, $Y_1+Y_2=\pm 2,\ Y_1-Y_2=0 \Rightarrow T=\pm 2$

2. When $B_1 \neq B_2$, $Y_1+Y_2=0,\ Y_1-Y_2=\pm 2 \Rightarrow T=\pm 2$

Thus $T=\pm 2$, Continue (3): $\sum\nolimits_{\Lambda}P_{\Lambda}\cdot T \le \sum\nolimits_{\Lambda}P_{\Lambda}\cdot 2 =2 $.


5. **Show that the following quantum states and measurements of Alice and Bob violate the above CHSH inequality**. First, Alice and Bob share entangled qubits each of which in the following form:
$$
\frac{1}{\sqrt{2}} \left( \left|01\right> - \left|10\right> \right).
$$
Imagine a quantum state as in the above is shared by Alice (the 1st-qubit) and Bob (the 2nd-qubit). They are free to measure their own qubit with the following measurement settings:     
    * Alice measures with $\{\left|0\right>, \left|1\right>\}$ or $\{\left|+\right>, \left|-\right>\}$.  
    * Bob measures with $\{\cos(\theta) \left|0\right> + \sin(\theta) \left|1\right>, -\sin(\theta) \left|0\right> + \cos(\theta) \left|1\right> \}$ or $\{\sin(\theta)\left|0\right> + \cos(\theta)\left|1\right>, -\cos(\theta)\left|0\right> + \sin(\theta)\left|1\right>\}$. 
    * Plot the value of the following equations with respect to the value of $0 \le \theta \le 2\pi$, and show that there is a range of $\theta$ where the CHSH inequality is violated, namely, 
    
    $$
    E[X_1 Y_1] + E[X_1 Y_2] + E[X_2 Y_1] - E[X_2 Y_2] > 2
    $$
    You can solve it by running quantum simulator and change the value of $\theta$. 

```python
import numpy as np
import math
from qiskit import(
  QuantumCircuit,
  QuantumRegister,
  ClassicalRegister,
  execute,
  Aer)
from qiskit.visualization import plot_histogram

# make quantum circuit
def make_circuit(alice=0, bob=0, theta=0):
    alice_measure = ClassicalRegister(1, "alice")
    bob_measure = ClassicalRegister(1, "bob")
    eprAlice = QuantumRegister(1, "eprAlice")   # Alice's part of EPR pair
    eprBob = QuantumRegister(1, "eprBob")    # Bob's part of EPR pair
    
    circuit = QuantumCircuit(eprAlice, eprBob, alice_measure, bob_measure)
    # create 1/√2 (|01⟩−|10⟩)
    circuit.x(0)
    circuit.x(1)
    circuit.h(1)
    circuit.cx(1,0)
    if alice == 0:
        # alice |0⟩ |1⟩, do nothing
        pass
    elif alice == 1:
        # alice |+⟩ |-⟩, apply inverse of H rotation = H
        circuit.h(0)
        
    # Ry(a) -> apply [[cos(a/2), - sin(a/2)], [sin(a/2), cos(a/2)] 
    if bob == 0:
        # bob cosθ|0⟩+sinθ|1⟩ -sinθ|0⟩+cosθ|1⟩
        # Ry(-2θ)|0⟩ = cosθ|0⟩ + sinθ|1⟩
        circuit.ry(-2 * theta, 1)
    elif bob == 1:
        # bob sinθ|0⟩+cosθ|1⟩ -cosθ|0⟩+sinθ|1⟩ 
        # cos(π/2-θ)|0⟩+sin(π/2-θ)|1⟩ -sin(π/2-θ)|0⟩+cos(π/2-θ)|1⟩ 
        # Ry(π-2θ)|0⟩ = cos(π/2-θ)|0⟩ + sin(π/2-θ)|1⟩ = sinθ|0⟩+cosθ|1⟩
        circuit.ry(2 * theta - np.pi, 1)
    circuit.measure(0, alice_measure)
    circuit.measure(1, bob_measure)
    return circuit
```

```python
qc = make_circuit(0, 0, math.radians(22))
qc.draw(output="mpl")
```

```python
# calculate expectation
def E(shots, counts):
    return sum([np.power(-1, x+y)*counts.get(f"{x} {y}", 0)/shots for x in range(2) for y in range(2)])

# run on theta
def run(theta, backend=Aer.get_backend("qasm_simulator"), **kwargs):
    Es = {} # expected value
    for alice_choice in range(2):
        for bob_choice in range(2):
            circuit = make_circuit(alice_choice, bob_choice, theta)
            
            shots = 8192
            job = execute(circuit, backend, shots=shots, **kwargs)
            counts = job.result().get_counts()
            
            Es[f"{alice_choice}{bob_choice}"] = E(shots, counts)
    
    res_00 = Es["01"]+Es["10"]+Es["11"]-Es["00"]
    res_01 = Es["00"]+Es["10"]+Es["11"]-Es["01"]
    res_10 = Es["01"]+Es["00"]+Es["11"]-Es["10"]
    res_11 = Es["01"]+Es["10"]+Es["00"]-Es["11"]
    return {"00": res_00, "01": res_01, "10": res_10, "11": res_11}
```

```python
run(7/8*np.pi)
```

```python
import matplotlib.pyplot as plt
# simulation
p = np.arange(0, 360, 1)
all_four_results = []
for theta in p:
    all_four_results.append(run(math.radians(theta)))   
for minus_comp in ["00", "01", "10", "11"]:
    plt.plot(p, list(map(lambda t: t[minus_comp], all_four_results)), label=minus_comp)
plt.legend()
```
